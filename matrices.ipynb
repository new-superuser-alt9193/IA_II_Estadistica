```python
def VarCovMatriz(matriz):
  matrizT = matriz.T
  n = matriz.shape[0]
  vectorUno = np.full((n, 1),1)
  print(vectorUno.shape)
  return (1 / (n - 1)) * (matrizT.dot(matriz) - (1 / n) * (matrizT.dot(vectorUno)).dot(matrizT.dot(vectorUno).T))
```
